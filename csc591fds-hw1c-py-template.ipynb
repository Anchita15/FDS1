## Visualize data

```python
## Generate single plot of historgrams for all attributes, and show legend
```

```python
df = pd.read_csv('C:/Users/Anchita/data/Advertising.csv')  

# Plotting histograms for all attributes in a single plot
df.plot(kind='hist', bins=10, alpha=0.5, figsize=(10, 6), edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histograms of All Attributes')

# Show legend
plt.legend(title='Attributes', labels=df.columns)

# Display the plot
plt.show()
```
